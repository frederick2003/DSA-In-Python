# Default Dictionaries
Default dictionaries allow the assignment of a default value to keys that fo not exist.

The real benefit of these is that you do not need to provide logic to check whether keys are present or not in the dictionary. 

```python
from collections import defaultdict
```

Any data type of structure can be passed as an argument to the *defaultdict(...) constructor*. This allows the direct modification of the default values and does not require the value to be manually set. The example below sets a default value of an empty list for any added key to the defaultdict. You are able to call list methods *append()* and *extend()* directly without needing to initialise the value first.

```python
d = defaultdict(list)
```

```python
d["x"].append('Appending this to the default value type')
print(d["x"])
```

```python
d["x"].extend(['add1', 'add2'])
print(d["x"])
```

If you try to access a non-existant key, non error is thrown, the default type is given, an empty list in this case.

```python
print(d["Non-existant-key"])
```

# K Most Frequent Elements Algorithm

```python
from collections import defaultdict
def topKFrequent(self, nums, k):
    d = defaultdict(int)

    # Build a frequency count of numbers
    for num in nums:
        d[num] += 1

    # Convert the frequency count to a list of tuples
    items = d.items()

    # Sort the tuples based on their count with the largest counts appearing first.
    sorted_items = sorted(items, key=lambda x: x[1], reverse=True)

    # Build a list, taking the numbers as the elements of the new list
    # The number at index 0 has the largest count, index 1 has the second largest and so on
    elements = [x[0] for x in sorted_items]

    # Return the k most frequent elements.
    return elements[:k]
```

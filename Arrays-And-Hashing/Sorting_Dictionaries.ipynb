# Sorting Dictionaries
You can sort a dictionary by using either its value or it's key. 

## Sorting a Dictionary by its Value.

Using a lambda function.

```python
d = {'banana' : 2, 'apple' : 3, 'watermelon' : 1}
print(d)
```

```python
# Sort in ascending order.
sorted_ascending = {k: v for k, v in sorted(d.items(), key=lambda item: item[1])}
```

```python
print(sorted_ascending)
```

```python
# Now lets sort in descending order
sorted_descending = {k: v for k,v in sorted(d.items(), key=lambda item:item[1], reverse=True)}
```

```python
print(sorted_descending)
```

## Sorting a Dictionary by its Key

### Method 1 

Sorting using a lambda function.

```python
sort_by_key = {3 : 'banana', 2 : 'apple', 1 : 'watermelon'}
print(sort_by_key)
```

```python
sort_asc = {k:v for k,v in sorted(sort_by_key.items(), key=lambda item:item[0])}
```

```python
print(sort_asc)
```

Now lets sort keys in descending order by passing the reverse=True parameter to the *sorted()* method.

```python
sort_desc = {k:v for k,v in sorted(sort_by_key.items(), key=lambda item:item[0], reverse=True)}
```

```python
print(sort_desc)
```

```python
from collections import defaultdict
nums = [4,1,-1,2,-1,2,3]
freq_table = defaultdict(int)
for num in nums:
    freq_table[num] += 1

descending = {k:v for k,v in sorted(freq_table.items(), key=lambda item: (-item[1], item[0]))}
print(descending)
```

```python
elements = list(descending.keys())
```

*ord(char)* converst a character to its unicode coding. *ord(char) - ord('a')* gives that chars order in the alphabet where *ord('a') - ord('a') = 0* and *ord('z') - ord('a')* = 25

```python
def groupAnagrams(strs):
    
    if len(strs) == 1:
        return [[strs[0]]]
    """
    The general idea is to create a dict where the key is a tuple.
    The tuple will have a count of the each letter where the index 
    corresponds to that letters position in the alphabet. 
    """

    # go over all words in the list
        # go over every char in the word
            # create a tuple representation as described above. 

    # Define the HashMap
    mapping = defaultdict(list)

    # Iterate over the words in the list
    for word in strs:
        # Create a list of same size as the alphabet and intialise all entries with 0.
        alph_encoding = [0] * 26

        # Iterate over chars in the current word                
        for char in word:

            # Get the alphabet position of the character
            pos = ord(char) - ord("a")

            # Increment the char count at the array position of that chars position
            alph_encoding[pos] += 1

        # convert the list to an immutable DS --> tuple
        alph_encoding_tup = tuple(alph_encoding)

        # Add entry to the dict where the key is the tuple and the value is the word
        mapping[alph_encoding_tup].append(word)

    return list(mapping.values())
```

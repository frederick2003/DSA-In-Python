```python
def isValid(self, s):
    # 1. Open brackets are closed by the associated closing bracket

    # 2. Open brackets must be closed in the correct order

    # 3. Every close bracket has a corresponding open bracket

    brackets = {"(" : ")", "{" : "}", "[" : "]"}

    # If open bracket 
        # add to the stack.

    # If closed bracket
        # Check if associated open bracket is on top of the stack

    stack = []
    for brac in s:

        if brac in brackets:
            stack.append(brac)
        else:
            if len(stack) == 0:
                return False 
                   
            cur = stack.pop()
            if brackets[cur] != brac:
                return False
    
    if len(stack) != 0:
        return False
    return True
```

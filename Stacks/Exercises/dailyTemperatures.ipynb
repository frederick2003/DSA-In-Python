```python
from collections import defaultdict

def dailyTemperatures(temperatures):
    """
    :type temperatures: List[int]
    :rtype: List[int]
    """
    answer = [0] * len(temperatures)
    stack = []
    n = len(temperatures)

    for i in range(n):
        while stack and temperatures[i] > temperatures[stack[-1]]:
            index = stack.pop()
            answer[index] = i - index
        stack.append(i)
    return answer
```

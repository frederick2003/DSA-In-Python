```python
def decodeString(s):
    # How to approach this question
    # This question is all about managing the logic within a sequence of bracks
    # The innermost bracket needs to be dealt with first.
    # This could be done recursively or by using a stack.

    # Add tokens to the stack until an open close bracket sequence is encountered.

    stack = []
    for token in s:
        if token != "]":
            stack.append(token)
        else:
            # Evaluate the substring
            substring = ""
            while stack[-1] != "[":
                substring = stack.pop() + substring
            stack.pop()

            # Evaluate the multiplier
            mult = ""
            while stack and stack[-1].isdigit():
                mult = stack.pop() + mult
            
            # Add the evaluated expression
            stack.append(substring * int(mult))
            
    answer = ""
    for token in stack:
        answer = answer + token
    return answer
```

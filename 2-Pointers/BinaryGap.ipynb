```python
def binaryGap(n):
    """
    :type n: int
    :rtype: int
    """

    # Left and right pointer
    # left = left
    # right = left + 1


    # when left and right point to 1's 
    # compare right - left with the global longest distance

    n_bin = format(n, "b")
    longest_dist = 0

    if len(n_bin) == 1:
        return 0
    
    left, right = 0, 1
    while right < len(n_bin):
        # if there are adjacent 1's calculate their dist and progress
        if n_bin[left] == "1" and n_bin[right] == "1":
            longest_dist = max(longest_dist, right - left)
            print(n_bin[left:right+1])
            print(f"left = {left}, right = {right}")
            left = right
            right += 1 

        # if left still doesn't point to a 1 progress both pointers
        elif n_bin[left] != "1":
            left += 1
            right += 1
        
        # if left points to 1 and and right points to 0
            # Progress only the right pointer
        elif n_bin[left] == "1" and n_bin[right] == "0":
            right += 1
    return longest_dist
```
